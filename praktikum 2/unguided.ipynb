# **4. Drop Missing Value Berdasarkan Baris, Kolom, Imputasi dengan Mean, Modus, dan Median**

A. Berdasarkan Baris

```python
data_baris = data.copy(deep=True)
drop_baris = data_baris.dropna(axis=0)
```

B. Berdasarkan Kolom

```python
data_kolom = data.copy(deep=True)
drop_kolom = data_kolom.dropna(axis=1)
```

C. Berdasarkan imputasi dengan mean

```python
data_mean = data.copy(deep=True)
data_mean['Time_taken'] = data_mean['Time_taken'].fillna(data_mean['Time_taken'].mean())
```

```python
data_mean.isna().sum()
```

D. Berdasarkan imputasi dengan modus

```python
data_modus = data.copy(deep=True)
data_modus['Time_taken'] = data_modus['Time_taken'].fillna(data_modus['Time_taken'].mode()[0])
```

```python
data_modus.isna().sum()
```

E. Berdasarkan imputasi dengan Modus

```python
data_median = data.copy(deep=True)
data_median['Time_taken'] = data_median['Time_taken'].fillna(data_median['Time_taken'].median())
```

```python
data_median.isnull().sum()
```
